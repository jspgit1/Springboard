```python
# load iris.csv
iris =  pd.read_csv('~/Springboard/unit5/iris.csv')
```

```python
# dataset practice
print(type(iris))
print(iris.shape)
print(iris.columns)
print(iris.iloc[0:10,:])
print(len(iris['species']))
#print(iris['species'][:])
```

```python
print('Max value in column sepal_length', iris['sepal_length'].max())
print('Mini value in column sepal_length', iris['sepal_length'].min())
print('Median value in column sepal_length', iris['sepal_length'].median())
print('standard deviation of each column', iris.std())
```

```python
#Json and dict format 
iris_json = iris.to_json()
print(iris_json)
iris_dict = iris.to_dict()
#print('\ndictionary format:',iris_dict)
```

```python
# plot
#import matplotlib.pyplot as plt
iris.plot(x='sepal_length', y='sepal_width',kind='scatter')
plt.show()
```

```python
# Using sepal_length and sepal_width features, we can distinguish 
# Setosa flower from others
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='species', height=4).map(plt.scatter,'sepal_length','sepal_width')\
    .add_legend()
plt.show()

```
